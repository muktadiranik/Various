```python
array = [1, 3, 5, 7, 9, 11, 13]

def search(array, target):
    for i in range(0, len(array) - 1):
        if array[i] == target:
            return i
    return -1

print(search(array, 5))

def binary_search(array, target):
    array = sorted(array)

    left = 0
    right = len(array) - 1

    while left <= right:
        middle = (left + right) // 2

        if target == array[middle]:
            return middle

        elif target < array[middle]:
            right = middle -1

        elif target > array[middle]:
            left = middle + 1

    return -1

print(binary_search(array, 5))

def binary_search_recursive(array, target, left, right):
    if left > right:
        return -1
    
    middle = (left + right) // 2

    if target == array[middle]:
        return middle
    
    if target < array[middle]:
        return binary_search_recursive(array, target, left, middle - 1)
    
    if target > array[middle]:
        return binary_search_recursive(array, target, middle + 1, right)
    
print(binary_search_recursive(array, 5, 0, len(array) - 1))
```

```python
# Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_at_ending(self, data):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            return
        
        current_node = self.head

        while current_node.next:
            current_node = current_node.next

        current_node.next = new_node

    def insert_at_position(self, data, position):
        new_node = Node(data)

        # inserting at the beginning
        if position == 0:
            new_node.next = self.head
            self.head = new_node
            return
        
        current_node = self.head
        count = 0

        # traverse to (position - 1) for 0 based index
        while current_node.next and count < position - 1:
            current_node = current_node.next
            count += 1

        # if position is not valid or out of range
        if not current_node:
            return
        
        # insert in the position
        new_node.next = current_node.next
        current_node.next = new_node


    def print_list(self):
        current_node = self.head

        while current_node:
            print(current_node.data, end=" -> ")
            current_node = current_node.next
        
        print(None)

    def delete(self, data):
        current_node = self.head

        if current_node and current_node.data == data:
            self.head = current_node.next
            return

        previous_node = None

        while current_node and current_node.data != data:
            previous_node = current_node
            current_node = current_node.next

        if current_node is None:
            return

        previous_node.next = current_node.next 

    def reverse_print_list(self):
        previous_node = None
        current_node = self.head

        while current_node:
            # store next
            next_node = current_node.next
            # reverse pointer
            current_node.next = previous_node
            # move previous forward
            previous_node = current_node
            # move current forward
            current_node = next_node

        self.head = previous_node

        self.print_list()
            


linked_list = LinkedList()

for i in range(5):
    linked_list.insert_at_beginning(i)

linked_list.print_list()

for i in range(5, 0, -1):
    linked_list.insert_at_ending(i)

linked_list.print_list()

linked_list.insert_at_position(-1, 6)

linked_list.print_list()

linked_list.delete(0)

linked_list.print_list()

linked_list.reverse_print_list()        
```

```python
# Selection Sort
from random import randint

array = [randint(1, 100) for i in range(1, 10)]

print(array)

def selection_sort(array):
    length = len(array)

    for i in range(length):
        minimum_value_index = i

        for j in range(i + 1, length):
            if array[minimum_value_index] > array[j]:
                minimum_value_index = j
        
        array[minimum_value_index], array[i] = array[i], array[minimum_value_index]
    
    return array

print(selection_sort(array))

def bidirectional_selection_sort(array, ascending=True):
    left_pointer = 0
    right_pointer = len(array) - 1

    while left_pointer < right_pointer:
        minimum_value_index = left_pointer
        maximum_value_index = right_pointer - 1

        for i in range(left_pointer, right_pointer + 1):
            if array[minimum_value_index] > array[i]:
                minimum_value_index = i
            if array[maximum_value_index] < array[i]:
                maximum_value_index = i

        array[left_pointer], array[minimum_value_index] = array[minimum_value_index], array[left_pointer]

        if maximum_value_index == left_pointer:
            maximum_value_index = minimum_value_index

        array[right_pointer], array[maximum_value_index] = array[maximum_value_index], array[right_pointer]

        left_pointer += 1
        right_pointer -= 1

    if not ascending:
        array.reverse()

    return array

print(bidirectional_selection_sort(array, True))
print(bidirectional_selection_sort(array, False))
```

```python
# Quick Sort
from random import randint


array = [randint(1, 100) for i in range(10)]
print(array)


def quick_sort(array):
    if len(array) <= 1:
        return array

    middle_index = len(array) // 2
    middle_index_value = array[middle_index]

    left_portion = [x for x in array if x < middle_index_value]
    middle_portion = [x for x in array if x == middle_index_value]
    right_portion = [x for x in array if x > middle_index_value]

    return quick_sort(left_portion) + middle_portion + quick_sort(right_portion)


print(quick_sort(array)) 
```

```python
# Hash Table
class HashTable:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(self.size)]

    def hash_function(self, key):
        return hash(key) % self.size
    
    def set(self, key, value):
        index = self.hash_function(key)

        # check if key exists
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)
                return

        # otherwise insert
        self.table[index].append((key, value))

    def get(self, key):
        index = self.hash_function(key)

        for k, v in self.table[index]:
            if k == key:
                return v
            
        return None
    

hash_table = HashTable()

hash_table.set("name", "Jhon")
hash_table.set("age", 30)

print(hash_table.get("age"))
```
