```python
import itertools
import string


characters = string.ascii_lowercase + string.ascii_letters + string.ascii_uppercase + string.digits + string.punctuation
count = 0

for length in range(8, 128):
    for p in itertools.permutations(characters, length):
        print(p)
        count += 1
        if count >= 1000000:
            break


for length in range(8, 128):
    for c in itertools.combinations(characters, length):
        print(c)
        count += 1
        if count >= 1000000:
            break
```

```python
characters = string.ascii_lowercase + string.ascii_uppercase + string.digits
count = 0

for length in range(8, 128):
    for p in itertools.product(characters, repeat=length):
        print(p)
        count += 1
        if count >= 1000000:
            break
```

```python
"""
Given an array of integers representing durations in minutes, 
return the number of pairs i, ji, j where i < ji < j and the sum of durations at these indices equals a full week. 
A full week is defined as a time duration that is an exact multiple of 10080 minutes. 
For example, 1 week is 10080 minutes, 2 week in 20160 minutes, 3 weeks is 30240 minutes, and so on. 

Example 1: 
input: durations = [5040, 5040, 6000, 10080, 10080] 
output: 2 
Explanation: the pairs of indices that form a full week are (0, 1) and (3, 4). 

Example 2: 
input: durations = [30240, 20160, 10080, 5] 
output: 3 
Explanation: the pairs of indices that form a full week are (0, 1), (0, 2), and (1, 2) 

constraints: 
1 <= durations.length <= 5 * 10^5 
1 <= durations[i] <= 10^9
"""
from collections import defaultdict


def solution(k: [int]) -> int: # type: ignore
    W = 10080  # 7 * 24 * 60
    freq = defaultdict(int)
    res = 0

    for d in k:
        r = d % W
        complement = (W - r) % W
        res += freq[complement]
        freq[r] += 1

    return res


if __name__ == "__main__":
    test_cases = [
        [6000, 12000, 10080],
        [3060, 2040, 15000, 10080, 4080],
        [3060, 6080, 10060],
    ]

    for i, test_case in enumerate(test_cases, 1):
        print(f"Test Case {i}: Input: {test_case} => Output: {solution(test_case)}")
```
