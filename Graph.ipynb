```python
# Depth First Search
def depth_first_search(graph, node, visited=None):
    if visited is None:
        visited = set()

    if node not in visited:
        print(node, end="\t")
        visited.add(node)

        for neighbor in graph[node]:
            depth_first_search(graph, neighbor, visited)


depth_first_search(graph, "A")
```

```python
from collections import deque


graph = {
    "A": ["B", "C"],
    "B": ["D"],
    "C": ["E"],
    "D": [],
    "E": []
}

def breadth_first_search(graph, start):
    visited = set()
    queue = deque([start])

    visited.add(start)

    while queue:
        node = queue.popleft()
        print(node, end="\t")

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)


breadth_first_search(graph, "A")
```

```python
# Dijkstra Algorithm
"""
1. start from a source node
2. assign distance = 0 for start node and distance = inf for other nodes
3. always pick the node with smallest distance
4. update neighbors if a shorter path is found
5. repeat until all nodes are visited
"""

graph = {
    "A": [("B", 1), ("C", 4)],
    "B": [("C", 2), ("D", 5)],
    "C": [("D", 1)],
    "D": []
}


import heapq


def dijkstra(graph, start):
    distances = {node: float("inf") for node in graph}
    distances[start] = 0

    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        # skip if we already found a better path
        if current_distance > distances[current_node]:
            continue

        # check neighbors
        for neighbor, weight in graph[current_node]:
            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances


dijkstra(graph, "A")
```

```python
# Dijkstra Algorithm
"""
1. start from a source node
2. assign distance = 0 for start node and distance = inf for other nodes
3. always pick the node with smallest distance
4. update neighbors if a shorter path is found
5. repeat until all nodes are visited
"""

"""
I have a map of cities(nodes), and roads between them(edges). Each road has a cost like distance or cost(weights).
"""

"""
Start from one city and find the shortest way to every other city.
"""

graph = {
    "A": [("B", 1), ("C", 4)],
    "B": [("C", 2), ("D", 5)],
    "C": [("D", 1)],
    "D": []
}

import heapq


def dijkstra(graph, start):
    # I do not know how far anything is, so everything = infinity
    # Start with guesses
    distances = {node: float("inf") for node in graph}

    # Initiate a distionary which will store the routes
    # Initiate the distionary with None value for every node or places or cities
    previous = {node: None for node in graph}

    # Start point is special because I am there
    # So start point = 0
    distances[start] = 0

    # This is a priority queue or magic bag
    # It always gives the closest place first
    priority_queue = [(0, start)]

    # Start exploring
    # Keep going until no places left
    while priority_queue:
        # Pick the closest place
        # Take the cheapest/closest place from the magic bag
        current_distance, current_node = heapq.heappop(priority_queue)

        # Skip bad path
        # If already found a better way, ignore this one
        if current_distance > distances[current_node]:
            continue

        # Check the current nodes neighbors and their distance or cost or weight
        # Check all nearby places or cities
        for neighbor, weight in graph[current_node]:
            # Calculate new distance
            distance = current_distance + weight
            # If found a better or closest path, update the distance and remember how I got there
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous[neighbor] = current_node # This is like saying "To reach B, I came from A"

                # Put it back in the magic bag so that it can be explored later
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances, previous


dijkstra(graph, "A")
```
