```python
import json

data = [
    {
    "name": "John",
    "age": 30,
    "city": "New York",
    }, 
    {
    "name": "Jane",
    "age": 25,
    "city": "San Francisco",
    }
]

data = json.dumps(data)
print(data)

with open(file="data.json", mode="w") as file:
    file.write(data)

with open(file="data.json", mode="r") as file:
    data = file.read()
    data = json.loads(data)
    print(data)
    print(data[0])
```

```python
import sqlite3
import json
from pathlib import Path

data = json.loads(Path("data.json").read_text())
print(data)

with sqlite3.connect("data.db") as connection:
    command = """
    CREATE TABLE IF NOT EXISTS Data (
        name TEXT, 
        age INTEGER, 
        city TEXT
    )"""
    connection.execute(command)
    connection.commit()

with sqlite3.connect("data.db") as connection:
    command = "INSERT INTO Data VALUES (?, ?, ?)"

    for item in data:
        connection.execute(command, tuple(item.values()))
    connection.commit()
```

```python
with sqlite3.connect("data.db") as connection:
    command = "SELECT * FROM Data"
    cursor = connection.execute(command)

    for row in cursor:
        print(row)


with sqlite3.connect("data.db") as connection:
    command = "SELECT * FROM Data"
    cursor = connection.execute(command)
    items = cursor.fetchall()
    print(items)
```

```python
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.image import MIMEImage
import smtplib
from pathlib import Path


message = MIMEMultipart()
message["from"] = "John Doe"
message["to"] = "muktadiranik@gmail.com"
message["subject"] = "Test email"

message.attach(MIMEText("This is the body of the email", "plain"))
message.attach(MIMEImage(Path("image.png").read_bytes()))

try:
    with smtplib.SMTP("smtp.gmail.com", 587) as smtp:
        smtp.ehlo()
        smtp.starttls()
        smtp.login("muktadiranik@gmail.com", "i.me.myself.okay.bye")
        smtp.send_message(message)
        print("Email sent!")
except Exception as e:
    print(e)
    print("Email failed to send.")
```

```python
import subprocess


try:
    task = subprocess.run(
        ["python", "3.py", "-a", "-b", "-c"], 
        shell=True, 
        capture_output=True, 
        text=True, 
        check=True
    )

    print(task)
    print(task.args)
    print(task.returncode)
    print(task.stdout)
    print(task.stderr)
except subprocess.CalledProcessError as e:
    print(e)
    print(e.args)
    print(e.stdout)
    print(e.stderr)
    print(e.returncode)
```
