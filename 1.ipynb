```python
# Class: Blueprint for creating objects
# Objects: Instance of a class
# Method: Function inside a class

# Class: Human
# Objects: Jhon, Alice
# Methods: talk(), walk()

class Point:
    default_color = "red"

    def __init__(self, x, y):
        self.x = x
        self.y = y

    @classmethod
    def zero(cls):
        return cls(5, 10)

    def draw(self):
        print(f"Point ({self.x}, {self.y})")

    def __str__(self):
        return f"Point ({self.x}, {self.y})"


point = Point(1, 2)
print(point.x, point.y)
point.draw()
print(type(point))
# Instance vs Class Attributes
print(isinstance(point, Point))
print(Point.default_color)
print(point.default_color)
point.default_color = "blue"
print(point.default_color)
point = point.zero()
print(point.x, point.y)
# Magic Methods
print(point.__dict__)
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __eq__(self, value):
        return self.x == value.x and self.y == value.y
    
    def __gt__(self, value):
        return (self.x > value.x) and (self.y > value.y)

    def __add__(self, value):
        return (self.x + value.x, self.y + value.y)
    
point_1 = Point(1, 2)
point_2 = Point(1, 2)
print(point_1 == point_2)
point_1 = Point(3, 4)
point_2 = Point(1, 2)
print(point_1 > point_2)
print(point_1 + point_2)
```

```python
class TagCloud:
    def __init__(self):
        self.__tags = {}

    def add(self, tag):
        self.__tags[tag.lower()] = self.__tags.get(tag.lower(), 0) + 1

    def __getitem__(self, tag):
        return self.__tags.get(tag.lower(), 0)
    
    def __setitem__(self, tag, count):
        self.__tags[tag.lower()] = count

    def __len__(self):
        return len(self.__tags)
    
    def __iter__(self):
        return iter(self.__tags)

cloud = TagCloud()
cloud.add("python")
cloud.add("Python")
cloud.add("PYTHON")
print(cloud)
print(cloud["python"])
print(cloud.__dict__)
```

```python
class InvalidOperationError(Exception):
    pass

class Stream:
    def __init__(self):
        self.opened = False

    def open(self):
        if self.opened:
            raise InvalidOperationError("Stream is already opened.")
        self.opened = True
        print("Stream opened.")

    def close(self):
        if not self.opened:
            raise InvalidOperationError("Stream is already closed.")
        self.opened = False
        print("Stream closed.")


class FileStream(Stream):
    def read(self):
        print("Reading data from file stream.")

class NetworkStream(Stream):
    def read(self):
        print("Reading data from network stream.")


file_stream = FileStream()
file_stream.open()
file_stream.read()

network_stream = NetworkStream()
network_stream.open()
network_stream.read()
```

```python
from abc import ABC, abstractmethod

# Good example of inheritance
class InvalidOperationError(Exception):
    pass


class Stream(ABC):
    def __init__(self):
        self.opened = False

    def open(self):
        if self.opened:
            raise InvalidOperationError("Stream is already opened.")
        self.opened = True
        print("Stream opened.")

    def close(self):
        if not self.opened:
            raise InvalidOperationError("Stream is already closed.")
        self.opened = False
        print("Stream closed.")

    @abstractmethod
    def read(self):
        pass



class FileStream(Stream):
    def read(self):
        print("reading data from file")


class NetwordStream(Stream):
    def read(self):
        print("reading data from network")



file_stream = FileStream()
file_stream.read()
print(file_stream.opened)
file_stream.open()

network_stream = NetwordStream()
network_stream.read()
print(network_stream.opened)
network_stream.open()
```

```python
from abc import ABC, abstractmethod

class UIControl(ABC):
    @abstractmethod
    def draw(self):
        print("abstract draw method")
        

class TextBox(UIControl):
    def draw(self):
        print("TextBox draw method")
        super().draw()


class DropDownList(UIControl):
    def draw(self):
        print("DropDownList draw method")
        super().draw()
    
def draw(controls):
    for control in controls:
        control.draw()
    

textbox = TextBox()
textbox.draw()

dropdownlist = DropDownList()
dropdownlist.draw()

draw([textbox, dropdownlist])
```
