```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.children = []

    def add_child(self, child):
        self.children.append(child)

    def __repr__(self, level=0):
        result = "\t" * level + self.value + "\n"
        for child in self.children:
            result += child.__repr__(level + 1)
        return result
```

```python
root = TreeNode("root")

laptop = TreeNode("laptop")
laptop.add_child(TreeNode("mac"))
laptop.add_child(TreeNode("windows"))

root.add_child(laptop)

phone = TreeNode("phone")
phone.add_child(TreeNode("android"))
phone.add_child(TreeNode("ios"))

root.add_child(phone)

print(root)
```

```python
def depth_first_traversal(node, level=0):
    print("\t" * level + node.value)
    for child in node.children:
        depth_first_traversal(child, level + 1)

depth_first_traversal(root)
```
